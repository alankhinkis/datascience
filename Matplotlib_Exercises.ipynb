Conduct a Bivariate Analysis for `"Protein"` and `"Chol"`. What is familar about this compared to the tutorial? What is another way to deal with overplotting other than the Hex plot? Create two data visualizations, one that shows the overplotting, another that resolves this issue. After you are done, play around with the sns.kdeplot function or try to create a 3D graph to reduce the effects of overplotting.

```python
# YOUR CODE HERE
#Scatter Plot
x = data["Protein"]
plt.xlabel("Protein")
y = data["Chol"]
plt.ylabel("Cholesterol")
plt.title("Scatter plot of Protein vs. Cholesterol")
plt.scatter(x, y, color='blue', marker='x')
plt.show()
#Scatter Plot with transparency
plt.scatter(x, y, color='blue', marker='x', alpha=0.3)
plt.show()
#Hexbin Plot
data.plot.hexbin(x='Protein', y='Chol', gridsize=25, cmap='Blues')

# 3D Plot
protein_bins = np.linspace(data['Protein'].min(), data['Protein'].max(), 20)
chol_bins = np.linspace(data['Chol'].min(), data['Chol'].max(), 20)

x_bins = np.digitize(x, protein_bins)
y_bins = np.digitize(y, chol_bins)

bin_counts = pd.DataFrame({'x_bins': x_bins, 'y_bins': y_bins})
counts = bin_counts.groupby(['x_bins', 'y_bins']).size().reset_index(name='count')

# Convert bin numbers back to bin centers for plotting
counts['x_center'] = protein_bins[counts['x_bins'] - 1] + (protein_bins[1] - protein_bins[0])/2
counts['y_center'] = chol_bins[counts['y_bins'] - 1] + (chol_bins[1] - chol_bins[0])/2

# 3D scatter
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(counts['x_center'], counts['y_center'], counts['count'],
           c=counts['count'], cmap='viridis', s=30)

ax.set_xlabel('Protein')
ax.set_ylabel('Cholesterol')
ax.set_zlabel('Count')
plt.show()
```

Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.¶


```python
def remove_char(old_string, n):
    return old_string[:n] + old_string[n+1:]
    

remove_char("sigma", 2)
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
    if len(old_string) < 2:
        return old_string
    new_string = old_string[-1] + old_string[1:-1] + old_string[:1]
    return new_string

```

Test the function with three different parameters


```python
# YOUR CODE HERE

print(swap_chars("bomboclat"))
print(swap_chars("I love KTP"))
print(swap_chars("WE yes WE are getting internships"))
```
